```python
data = xr.open_dataset(files[4])
var = 'tcc'
if var == 'q' or var == 'r':
    data = data.sel(level = 1000)
```

```python
ax = sns.heatmap( data.mean('time')[var].values)
ax.set_title("MEAN {}".format(var))
```

```python
ax = sns.heatmap(data.var('time')[var].values)
ax.set_title('Variance {}'.format(var))
```

```python
ax = sns.heatmap(data.std('time')[var].values)
ax.set_title('Standard Deviation {}'.format(var))
```

```python
ax = sns.heatmap(data.median('time')[var].values)
ax.set_title('Median {}'.format(var))
```

```python
ax = sns.heatmap(data.min('time')[var].values)
ax.set_title('MIN {}'.format(var))
```

```python
ax = sns.heatmap(data.max('time')[var].values)
ax.set_title('MAX {}'.format(var))
```
