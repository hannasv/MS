```python
import os
import glob
import json

import matplotlib.pyplot as plt
import xarray as xr
import numpy as np

def get_array(data, key = 'loss'):
    return [data[key][k] for k in data[key].keys()]

test_models = glob.glob('/home/hanna/EX3_Results/*')

fig, axes = plt.subplots(2, 1, figsize = (6.1023622, 6.1023622*.8), sharex=True)
num_converged_models = 0
min_name = ''
min_val = 1000
for path in test_models[:-1]:
    try:
        with open(path + '/history.json') as json_file:
            data = json.load(json_file)

        model = path.split('/')[-1]
        num_converged_models +=1
        # summarize history for accuracy
        axes[0].plot(get_array(data, key = 'loss'), label = model)
        axes[1].plot(get_array(data, key = 'val_loss'), label = model)
        
        if np.min(get_array(data, key = 'val_loss') ) < min_val:
            min_name = path
            min_val = np.min(get_array(data, key = 'val_loss') )
    except FileNotFoundError:
        print('file not found for model {}'.format(path))

axes[0].set_ylabel('Training Loss')
axes[1].set_ylabel('Validation Loss')
axes[1].set_xlabel('Epoch')
#axes[0].set_xlabel('Epoch')

axes[0].legend(ncol = 1, frameon = False, loc='upper right', # bbox_to_anchor=(x0, y0, width, height),
            edgecolor='w', fontsize='small', framealpha=0.5, #frameon = False,
            borderaxespad=0)

#plt.legend(ncol = 1, frameon = True, loc='upper right') #bbox_to_anchor=(0.5, 1.8))
#axes[0].grid(True)
#axes[1].grid(True)
#plt.legend(['train', 'test'], loc='upper left')
#plt.subplots_adjust(wspace = 0.2, hspace = 0.3, top=0.9, bottom=0.1, left = 0.14, right = .95)
plt.subplots_adjust(wspace = 0.3, hspace = 0.2, top= 0.95, bottom= 0.1, left= 0.15, right= 0.95)
#plt.savefig(os.path.join('/home/hanna/MS-thesis/python_figs/', 'test_epoch_loss.pdf'))
```

```python
dummy_test_mae = 0.011
dummy_test_mse = 0.982

#date = '2014-01-01' # burde sjekke hvordan skyene var da 

dummy_sequence_prediction = np.random.random((24, 81, 161)) 

def save_prediction(test_mae, test_mse, sequence_prediction):
    """ Save prediction """
    data_dict = {'tcc': (['time', 'latitude', 'longitude'], sequence_prediction)}

    longitude  = np.arange(-15.0, 25.0+0.25, step = 0.25)
    latitude   =  np.arange(30.0, 50.0+0.25, step = 0.25)
    seq_length = 24
    time = np.arange(seq_length)

    ds = xr.Dataset(data_dict,
                     coords={'longitude': (['longitude'], longitude),
                             'latitude': (['latitude'], latitude),
                             'time': (['time'], time)
                            })
    ds['mse_test']  = test_mse
    ds['mae_test']  = test_mae
    ds['date_seq']  = '2014-01-01'

    ds.to_netcdf('/home/hannasv/results_new/test_prediction_conv_lstm.nc')
    return 
```

```python
save_prediction(test_mae=dummy_test_mae, test_mse=dummy_test_mse, sequence_prediction=dummy_sequence_prediction)
```
