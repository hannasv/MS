```python
import matplotlib.style
import matplotlib as mpl
#mpl.style.use('classic')
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.plot(np.arange(10))
plt.xlabel(" Hei hallo ", fontsize = 20)
plt.ylabel(" Er du her ", fontsize = 20)
plt.title(" kkkkkkkkk ", fontsize = 15)
```

```python
mpl.rcParams['mathtext.fontset'] = 'custom'
mpl.rcParams['mathtext.it'] = 'Arial:italic'
mpl.rcParams['mathtext.rm'] = 'Arial'
```

```python
plt.plot(np.arange(10))
plt.xlabel(" Hei hallo ", fontsize = 20)
plt.ylabel(" Er du her ", fontsize = 20)
plt.title(" kkkkkkkkk ", fontsize = 15)
plt.ylabel("Median Population", fontname="Arial", fontsize=20)
plt.xlabel("Median Population", fontname="Georgia", fontsize=20)
```
